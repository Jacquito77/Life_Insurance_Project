```python
data = data.drop(['Medical_History_10','Medical_History_24','Medical_History_32'], axis=1)
```

```python
data = pd.get_dummies(data, columns=['Product_Info_2'], prefix = ['Product_Info_2'])
```

```python
data = data.drop(["Employment_Info_1", "Employment_Info_4", "Employment_Info_6", "Insurance_History_5", "Family_Hist_2", "Family_Hist_3", "Family_Hist_4", "Family_Hist_5", "Medical_History_1", "Medical_History_15", "Id"], axis=1)
```

```python
data.head()
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
data.columns.get_loc("Response")
```

```python
resp = data["Response"]
data.drop(labels=["Response"], axis=1,inplace = True)
data.insert(0, "Response", resp)
```

```python
X, y = data.iloc[:,1:-1],data.iloc[:,1]
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
from sklearn.model_selection import train_test_split

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
test = pd.read_csv("data/test.csv") # load the testing data
test['BMI_Age'] = test['BMI']*test['Ins_Age']
y_train = data.Response
X_train = data.drop(["Response"], axis=1)
```

```python
test.head()
```

```python
ids = test[['Id']]
test = test.drop(["Medical_History_10","Medical_History_24","Medical_History_32","Employment_Info_1", "Employment_Info_4", "Employment_Info_6", "Insurance_History_5", "Family_Hist_2", "Family_Hist_3", "Family_Hist_4", "Family_Hist_5","Medical_History_15", "Medical_History_1", "Id"], axis=1)
test = pd.get_dummies(test, columns=['Product_Info_2'], prefix = ['Product_Info_2'])
#test = test[data.columns]
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_reg.fit(X_train,y_train)

preds = xg_reg.predict(test)
```

```python
rmse = np.sqrt(mean_squared_error(ids, preds))
print("RMSE: %f" % (rmse))
```

```python
params = {"objective":"reg:linear",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3,
                    num_boost_round=50,early_stopping_rounds=10,metrics="rmse", as_pandas=True, seed=123)
```

```python
cv_results.head()
```

```python
xg_reg = xgb.train(params=params, dtrain=data_dmatrix, num_boost_round=10)
```
